```python
# Remove any leading '1,' or whitespace from the SIC codes
def clean_sic_codes(sic_str):
    if pd.isna(sic_str):
        return None
    # Split by comma in case there are multiple codes
    codes = [code.strip() for code in str(sic_str).split(',')]
    # Remove any '1' entries
    codes = [code for code in codes if code != '1' and code != '']
    # Join back as comma-separated string if multiple
    return ','.join(codes) if codes else None

# Apply to your SIC column
pay_gap['SicCodes'] = pay_gap['SicCodes'].apply(clean_sic_codes)
```

```python
# Function to map SIC codes to UK sections From A-U
def map_sic_section(sic_code):
    try:
        prefix = int(str(sic_code)[:2])
    except:
        return 'Other / Unknown'
    
    if 1 <= prefix <= 3:
        return 'Agriculture, Forestry and Fishing'
    elif 5 <= prefix <= 9:
        return 'Mining and Quarrying'
    elif 10 <= prefix <= 33:
        return 'Manufacturing'
    elif 35 <= prefix <= 35:
        return 'Electricity, Gas, Steam and Air Conditioning Supply'
    elif 36 <= prefix <= 39:
        return 'Water Supply; Sewerage, Waste Management and Remediation Activities'
    elif 41 <= prefix <= 43:
        return 'Construction'
    elif 45 <= prefix <= 47:
        return 'Wholesale and Retail Trade; Repair of Motor Vehicles and Motorcycles'
    elif 49 <= prefix <= 53:
        return 'Transportation and Storage'
    elif 55 <= prefix <= 56:
        return 'Accommodation and Food Service Activities'
    elif 58 <= prefix <= 63:
        return 'Information and Communication'
    elif 64 <= prefix <= 66:
        return 'Financial and Insurance Activities'
    elif 68 <= prefix <= 75:
        return 'Real Estate, Professional, Scientific and Technical Activities'
    elif 77 <= prefix <= 82:
        return 'Administrative and Support Service Activities'
    elif 84 <= prefix <= 88:
        return 'Public Administration and Defence; Compulsory Social Security'
    elif 90 <= prefix <= 93:
        return 'Arts, Entertainment and Recreation'
    elif 94 <= prefix <= 96:
        return 'Other Service Activities'
    elif 97 <= prefix <= 98:
        return 'Activities of Households as Employers; Undifferentiated Goods- and Services-Producing Activities of Households for Own Use'
    elif prefix == 99:
        return 'Activities of Extraterritorial Organisations and Bodies'
    else:
        return 'Other / Unknown'

# Apply the function to SIC code column
pay_gap['IndustrySection'] = pay_gap['SicCodes'].apply(map_sic_section)
```

```python
import re

# Example: full postcode cleaning
def extract_area(postcode):
    if pd.isna(postcode):
        return None
    postcode = postcode.upper().strip()
    # Extract first 1-2 letters at the start
    match = re.match(r'^([A-Z]{1,2})', postcode)
    return match.group(1) if match else None

pay_gap['PostcodeArea'] = pay_gap['PostCode'].apply(extract_area)
```

```python
def extract_postcode_area(pc):
    if pd.isna(pc):
        return np.nan
    
    match = re.match(r"^[A-Z]{1,2}", pc)
    return match.group(0) if match else np.nan

pay_gap['PostcodeArea'] = pay_gap['PostCode'].apply(extract_postcode_area)
```

```python
# Function to get region from address (case-insensitive)
def get_region_from_address(address):
    if pd.isna(address):
        return np.nan
    
    address_lower = address.lower()
    # Keyword-based mapping
     # Keyword-based mapping
    if any(keyword in address_lower for keyword in ['spain', 'madrid']):
        return 'Spain'
    elif any(keyword in address_lower for keyword in ['germany', 'munich']):
        return 'Germany'
    elif any(keyword in address_lower for keyword in ['switzerland', 'basel']):
        return 'Switzerland'
    elif any(keyword in address_lower for keyword in ['usa', 'boston', 'wilmington', 'new york', 'lexington', 'beltsville', 'maryland']):
        return 'USA'
    elif 'dublin' in address_lower or 'ireland' in address_lower:
        return 'Ireland'
    elif any(keyword in address_lower for keyword in ['mumbai', 'maharashtra', 'mahar']):
        return 'India'
    elif any(keyword in address_lower for keyword in ['norwich', 'wolverhampton', 'sheffield', 'buckshaw', 'lichfield', 'st h', 'london', 'slough', 'ryde', 'bath', 'swindon', 'st. john street', 'cardiff', 'scotland', 'gr']):  # gr -> Grangemouth or other Scottish addresses
        return 'UK'
    elif 'the hague' in address_lower or 'netherlands' in address_lower:
        return 'Netherlands'
    elif 'luxembourg' in address_lower:
        return 'Luxembourg'
    elif 'bermuda' in address_lower or 'hamilton' in address_lower:
        return 'Bermuda'
    elif 'brussels' in address_lower or 'belgium' in address_lower:
        return 'Belgium'
    elif 'gibraltar' in address_lower:
        return 'Gibraltar'
    elif 'france' in address_lower or 'paris' in address_lower:
        return 'France'
    elif 'stockholm' in address_lower or 'sweden' in address_lower:
        return 'Sweden'
    elif 'tokyo' in address_lower or 'japan' in address_lower:
        return 'Japan'
    elif 'sydney' in address_lower or 'australia' in address_lower:
        return 'Australia'
    else:
        return np.nan

# Only update Region where it's NaN
mask = pay_gap['Region'].isna()
pay_gap.loc[mask, 'Region'] = pay_gap.loc[mask, 'Address'].apply(get_region_from_address)

(pay_gap['Region'].isna().sum())
```
